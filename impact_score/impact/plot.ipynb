```python
from impact_score.impact.graph_plot import GraphPlotter
import seaborn as sns
from matplotlib import style as mpl_style
from matplotlib import pyplot as plt
from matplotlib import ticker as mpl_ticker
import time

match_id = 60206
gp = GraphPlotter(match_id)
```

```python
gp.base_plot(round_number=5)
```

```python
chosen_side = "atk"
color_dict = {"atk": "red", "def": "blue"}
title_dict = {"atk": "Attack", "def": "Defense"}

fig = plt.figure(figsize=(14, 7))
fig.tight_layout()
plt.rcParams.update({'font.size': 13})
# sns.set_context(rc={'patch.linewidth': 2.0})
# sns.set(font_scale=1.3)
# ax = sns.lineplot(data=gp.plot_data, x="Round time", y="Win_probability",ci=None,
#                 linewidth=1, zorder=0, color=color_dict[chosen_side])

plt.plot(gp.plot_data["Round time"], gp.plot_data["Win_probability"], zorder=0, color=color_dict[chosen_side])
ax = plt.gca()
ax.spines["left"].set_position(("data", -4))
ax.set_xlabel('Round time (s)', rotation=0, labelpad=10)
ax.set_ylabel('Win probability (%)', rotation=90, labelpad=20)
ax.grid(b=True, which="major", color="orange", linestyle="-", zorder=0, linewidth=.1, alpha=0.5)

formatter = mpl_ticker.FuncFormatter(lambda ms, x: time.strftime('%M:%S', time.gmtime(ms // 1000)))
ax.xaxis.set_major_formatter(formatter)

# ax.xaxis.labelpad = 10
# ax.yaxis.labelpad = 12
plt.title(f"{title_dict[chosen_side]} win probability over time")



# ax.lines[0].set_marker("o")
# ax.lines[0].set_markersize(9)

x_data = gp.plot_data["Round time"]
y_data = gp.plot_data["Win_probability"]
color_data = gp.plot_data["Colors"]
label_data = gp.plot_data["Tags"]
for round_time, win_probability, color, label in zip(x_data, y_data, color_data, label_data):
    if color is not None:
        ax.scatter(x=round_time, y=win_probability, color=color, edgecolor="w", marker="o", s=70, label=label)

plt.legend(framealpha=1, frameon=True)
#plt.axhline(y=9, color="black")
plt.show()
```
