```python
from impact_score.model.model_improvement.probability_calibration import ProbabilityCalibration
import matplotlib.pyplot as plt

pc = ProbabilityCalibration()
input_df = pc.create_multiple_gamestates("Loadout_diff", (-10000, 10000, 1000))
```

```python
variable_name = input_df["Feature"].unique()[0]
plt.figure(figsize=(15,7))
plt.xlabel(variable_name)
plt.ylabel("Probability")
plt.plot(input_df[variable_name], input_df["Probability"])
plt.grid(b=True, which='major', color='#666666', linestyle='-', linewidth=0.15)
plt.scatter(input_df[variable_name], input_df["Probability"], marker="o", color="red", s=20, edgecolors="black")
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:.0%}".format(x)))
plt.show()
```
