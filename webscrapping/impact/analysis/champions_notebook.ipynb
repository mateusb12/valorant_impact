```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("champions_analysis.csv")
df
```

```python
# Change the figsize
plt.figure(figsize=(10,6))
x_axis = "Fav Rifle"
y_axis = "Delta_sum (%)"
ax = sns.violinplot(x=x_axis, y=y_axis, data=df)
sns.stripplot(x=x_axis, y=y_axis, data=df,
              color="maroon", edgecolor="gray")

ax.set_xlabel("Favorite Rifle",fontsize=15)
ax.set_ylabel("Player Delta Impact",fontsize=15)
```
